```python
data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Evan"],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing"],
    "Age": [29, 35, 41, 28, 32],
    "Salary": [45000, 65000, 70000, 60000, 52000],
    "Experience_Years": [4, 10, 15, 3, 7]
}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
```
